```python
##Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##reading data

df=pd.read_csv("zomato.csv")
df.head()
```

```python
## analysing data
```

```python
df.shape
```

```python
df.columns
```

```python
##remove unnecessary columns
df=df.drop(['url', 'address','phone','dish_liked','reviews_list', 'menu_item'],axis=1)
df.head()
```

```python
df.info()
```

```python
## remove duplicates
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.info()
```

```python
df['rate'].unique()
```

```python
def change(value):

    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
```

```python
df['rate'] = df['rate'].apply(change)
df['rate'].head()
```

```python
##filling null values
df['rate'].fillna(df['rate'].mean(),inplace=True)
```

```python
df.info()
```

```python
## others columns don't have so many null values so i have to remove all the rows which contains the null values
df.dropna(inplace=True)
df.head()
```

```python
## rename the column 
df.rename(columns = {'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
## location and listed_in(city) both are same values, so i have to remove one column 

df= df.drop(['listed_in(city)'],axis=1)
df
```

```python
df.info()
```

```python
df['Cost2plates'].unique()
```

```python
df['Cost2plates']=df['Cost2plates'].str.replace(',','')
```

```python
df['Cost2plates'].unique()
```

```python
df['Cost2plates']=df['Cost2plates'].astype(np.int64)
```

```python
df.info()
```

```python
df
```

```python
rest_types=df['rest_type'].value_counts()
rest_types
```

```python
rest_types_lessthan_1000 = rest_types[rest_types<1000]
rest_types_lessthan_1000
```

```python
def handle_rest_values(i):
    if (i in rest_types_lessthan_1000):
        return 'others'
    else:
        return i 
```

```python
df['rest_type']=df['rest_type'].apply(handle_rest_values)
```

```python
df['rest_type'].value_counts()
```

```python
location1 = df['location'].value_counts()
location1
```

```python
location_lessthan_300 = location1[location1<300]
location_lessthan_300
```

```python
def location_lessthan300(i):
    if (i in location_lessthan_300):
        return 'others'
    else:
        return i
```

```python
df['location']=df['location'].apply(location_lessthan300)
```

```python
df['location'].value_counts()
```

```python
df.head()
```

```python
cuisines1=df['cuisines'].value_counts()
cuisines1
```

```python
cuisines_lessthan_100=cuisines1[cuisines1<100]
cuisines_lessthan_100
```

```python
def cuisines_lessthan100(i):
    if (i in cuisines_lessthan_100):
        return 'others'
    else:
        return i
```

```python
df['cuisines']=df['cuisines'].apply(cuisines_lessthan100)
df['cuisines'].value_counts()
```
